```python
import numpy as np 
from pandas import DataFrame
import seaborn as sns
%matplotlib inline

Index= ['aaa', 'bbb', 'ccc', 'ddd', 'eee']
Cols = ['A', 'B', 'C', 'D']
input = abs(np.random.randn(5, 4))
df = DataFrame(input, index=Index, columns=Cols)

sns.heatmap(df, annot=True)
```
