```python
match_dfs = []

for s in range(2012,2018):
    with open('Data/Match_info/'+str(s)+'.json', 'r') as f:
        data = json.load(f)
    
    for w in data:
        for m in data[w]:
            match_info = data[w][m]
            match_info['season'] = str(s)
            match_info['week'] = w
            match_info['match'] = m
            match_dfs.append(pd.DataFrame(match_info, index=[0]))
```

```python
super_rugby_data = pd.concat(match_dfs)
```

```python
super_rugby_data.groupby('season').size().reset_index()
```

```python
super_rugby_data.reset_index(drop=True, inplace=True)
```

```python
# handle missing data
```

```python
# containin - as a string.
super_rugby_data.isin(['-']).any()[super_rugby_data.isin(['-']).any() == True]
```

```python
# All of the are accuracy when their were no attempt for conversion, drop goal or penalty
super_rugby_data = super_rugby_data.replace('-', '0.0%')
```

```python
# containin Nan as a string.
super_rugby_data.isin(['NaN']).any()[super_rugby_data.isin(['NaN']).any() == True]
```

```python
# possentions of different sections of the field not available.
super_rugby_data = super_rugby_data.replace('NaN', '0.0%')
```

```python
super_rugby_data.columns
```

```python
# There are couple of columns with , in it as a thousand seperators e.g. (meters carried)
for c in super_rugby_data.columns:
    super_rugby_data[c] = super_rugby_data[c].str.replace(',', "")
```

```python
# convert relevent columns to int
col_int = ['carries1', 'carries2', 'clean_breaks1', 'clean_breaks2', 'conversions1',
           'conversions2', 'defenders_beaten1', 'defenders_beaten2','drop_goals1',
           'drop_goals2', 'kicks_in_play1', 'kicks_in_play2', 'lineouts_lost1', 
           'lineouts_lost2', 'lineouts_won1', 'lineouts_won2', 'mauls_won1', 
           'mauls_won2', 'meters_carried1', 'meters_carried2', 'missed_tackels1', 
           'missed_tackels2', 'offloads1', 'offloads2', 'passes1', 'passes2', 
           'penalties_conceded1', 'penalties_conceded2', 'penalty_goals1',
           'penalty_goals2', 'red_cards1', 'red_cards2', 'rucks_lost1', 'rucks_lost2', 
           'rucks_won1', 'rucks_won2', 'score1', 'score2', 'scrums_lost1', 
           'scrums_lost2', 'scrums_won1', 'scrums_won2', 'tackles1', 'tackles2', 
           'tries1', 'tries2', 'turnovers_conceded1', 'turnovers_conceded2', 
           'turnovers_won1', 'turnovers_won2', 'yellow_cards1', 'yellow_cards2', 'season']
```

```python
for c in col_int:
    super_rugby_data[c] =  super_rugby_data[c].astype(int)
```

```python
# floats are avaible % s, remove % sign and conver to float
col_float = ['conversion_accuracy1', 'conversion_accuracy2', 'drop_goal_accuracy1', 
             'drop_goal_accuracy2', 'overall_pos1', 'overall_pos2', 
             'penalty_goal_accuracy1', 'penalty_goal_accuracy2', 
             'team1_25', 'team1_40', 'team1_op_25', 'team1_op_40', 'team2_25', 
             'team2_40', 'team2_op_25', 'team2_op_40']
```

```python
for c in col_float:
    super_rugby_data[c] =  super_rugby_data[c].str.replace('%',"")
    super_rugby_data[c] =  super_rugby_data[c].astype(float)
```

```python
# names to string
col_names = ['team1', 'team2']
```

```python
for c in col_names:
    super_rugby_data[c] =  super_rugby_data[c].astype('category')
```

```python
# few transformations
super_rugby_data.week = super_rugby_data.week.replace({'week_':''}, regex = True)\
                            .astype(int)
super_rugby_data.match = super_rugby_data.match.replace({'match_':''}, regex = True)\
                            .astype(int)
```

```python
# Data is cleaned
super_rugby_data.head()
```

```python
# dealing NaNs created in the process
super_rugby_data.isnull().any()[super_rugby_data.isnull().any()==True]
```

```python
# all these NaNs can be treated as 0
col_NA = ['overall_pos1', 'overall_pos2', 'team1_25', 'team1_40', 'team1_op_25', 
          'team1_op_40', 'team2_25', 'team2_40', 'team2_op_25', 'team2_op_40']

for c in col_NA:
    super_rugby_data[c] = super_rugby_data[c].fillna(0)
```

```python
# 2012 data do not have possetion info fof all terretories. 
# Most useful is the overall possetion

super_rugby_data.overall_pos1 = super_rugby_data.team1_25+super_rugby_data.team1_40+\
                                super_rugby_data.team1_op_25+super_rugby_data.team1_op_40+\
                                super_rugby_data.overall_pos1

super_rugby_data.overall_pos2 = super_rugby_data.team2_25+super_rugby_data.team2_40+\
                                super_rugby_data.team2_op_25+super_rugby_data.team2_op_40+\
                                super_rugby_data.overall_pos2      
```

```python
super_rugby_data.drop(['team1_25', 'team1_40', 'team1_op_25', 'team1_op_40', 
                       'team2_25', 'team2_40', 'team2_op_25', 'team2_op_40'], axis=1, 
                     inplace = True)
```
