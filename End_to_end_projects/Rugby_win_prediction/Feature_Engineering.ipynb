# Feature Engineering

```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', 500)
```

```python
super_rugby = pd.read_feather('Data/super_rugby_data_cleaned')
```

```python
super_rugby.head()
```

```python
# get match id for reference
super_rugby = super_rugby.sort_values(['season', 'week', 'match']).reset_index(drop=True)
```

```python
super_rugby = super_rugby.reset_index()#.rename({'index':'match_id'}, axis=1)
```

```python
super_rugby = super_rugby.rename(columns = {'index':'match_id'})
```

```python
super_rugby.head()
```

```python
# indicate win_loss for match
def derive_result1(row):
    if row['score1'] >= row['score2']: return 1
    else: return 0

# match draws is indicated is as both teams winning
super_rugby['result1'] = super_rugby.apply(lambda row: derive_result1(row), axis=1)
super_rugby['result2'] = 1 - super_rugby['result1']
```

```python
# score difference as a feature
super_rugby['score_diff1'] = (super_rugby.score1 - super_rugby.score2)
super_rugby['score_diff2'] = (super_rugby.score2 - super_rugby.score1)
```

```python
super_rugby.head()
```

```python
# make dataframe long to get aggregates by teams
```

```python
col1 = []
col2 = []
common = []
for c in super_rugby.columns:
    if '1' in c:
        col1.append(c)
    elif '2' in c:
        col2.append(c)
    else:
        common.append(c)
```

```python
# standardized columns
s_col = [c[:-1] for c in col1]
```

```python
team1_df = super_rugby[common+col1]
```

```python
team1_df.columns = (common+s_col)
```

```python
team1_df.head()
```

```python
team1_df.shape
```

```python
team2_df = super_rugby[common+col2]
```

```python
team2_df.columns = (common+s_col)
```

```python
team2_df.head()
```

```python
team2_df.shape
```

```python
# indicate the team hosting the match
team1_df['host'] = 0
team2_df['host'] = 1
```

```python
super_rugby_long = pd.concat([team1_df, team2_df])
```

```python
super_rugby_long = super_rugby_long.sort_values(['season', 'team', 'week']).reset_index(drop=True)
```

```python
super_rugby_long.head()
```

```python
super_rugby_long.shape
```

```python
# match details
match_info = super_rugby_long[['match_id', 'season', 'week', 'team', 
                               'host', 'result']].copy()
```

```python
match_info.head()
```

```python
match_info.shape
```

```python
# aggregating to find value using last 5 matches
```

```python
avg_cols = ['carries', 'clean_breaks',
       'conversion_accuracy', 'conversions', 'defenders_beaten',
       'drop_goal_accuracy', 'drop_goals', 'kicks_in_play', 'lineouts_lost',
       'lineouts_won', 'mauls_won', 'meters_carried', 'missed_tackels',
       'offloads', 'overall_pos', 'passes', 'penalties_conceded',
       'penalty_goal_accuracy', 'penalty_goals', 'red_cards', 'rucks_lost',
       'rucks_won', 'score', 'scrums_lost', 'scrums_won', 'tackles',
       'tries', 'turnovers_conceded', 'turnovers_won', 'yellow_cards',
       'score_diff']

super_rugby_agg = super_rugby_long[['season', 'team']+avg_cols]
```

```python
super_rugby_agg.head(6)
```

```python
super_rugby_agg.shape
```

```python
super_rugby_agg = super_rugby_agg.groupby(['season', 'team']).rolling(5).mean()\
                    .reset_index(drop = True).drop(['season', 'team'], axis=1)
```

```python
super_rugby_agg.shape
```

```python
super_rugby_agg.head()
```

```python
# shifted match_id and host up
match_id = list(match_info.match_id[1:])+[None]
host = list(match_info.host[1:])+[None]

super_rugby_agg['match_id'] = match_id
super_rugby_agg['host'] = host
```

```python
super_rugby_agg['team'] = match_info.team
```

```python
super_rugby_agg = super_rugby_agg.dropna()
```

```python
super_rugby_agg.shape
```

```python
super_rugby_agg.head()
```

```python
# make sure teams only have their corresponding match_ids
super_rugby_agg = super_rugby_agg.merge(match_info[['match_id','host','team']], 
                                 on = ['match_id','host','team'])
```

```python
# lets get matchid which are not balanced
match_id_keep = super_rugby_agg.groupby('match_id').size()\
[super_rugby_agg.groupby('match_id').size() ==2].index
```

```python
super_rugby_agg = super_rugby_agg[super_rugby_agg.match_id.isin(match_id_keep)]
```

```python
super_rugby_agg.shape
```

```python
super_rugby_agg.head()
```

```python
# get seperate data frames for host and visit to take differences
host_df = super_rugby_agg[super_rugby_agg.host == 0].copy().\
            sort_values('match_id').drop(['match_id', 'host', 'team'], axis=1)
visit_df = super_rugby_agg[super_rugby_agg.host == 1].copy().\
            sort_values('match_id').drop(['match_id', 'host', 'team'], axis=1)
```

```python
host_df.head()
```

```python
visit_df.head()
```

```python
# columns after difference
diff_cols = host_df.columns
diff_cols = ['d_'+c for c in diff_cols]
```

```python
# for difference columns
# difference is better since denominators can zero
host_df_diff = pd.DataFrame(np.array(host_df)-np.array(visit_df))
visit_df_diff = pd.DataFrame(np.array(visit_df)-np.array(host_df))
host_df_diff.columns = diff_cols
visit_df_diff.columns = diff_cols
```

```python
feature_df = pd.concat([host_df_diff, visit_df_diff])
```

```python
feature_df.shape
```

```python
feature_df = feature_df.reset_index(drop = True)
feature_df.head()
```

```python
# get match_id, host, team columns
feature_df = super_rugby_agg[['match_id', 'host', 'team']].\
                sort_values(['host', 'match_id']).reset_index(drop = True).\
                join(feature_df, how = 'inner')
```

```python
feature_df.shape
```

```python
feature_df.head()
```

```python
# combine with match_info to get response variable
feature_df = match_info.merge(feature_df, on = ['match_id', 'host', 'team'])
```
